```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
y1=2*x
y2=3*x
```

```python
y1=2*x
```

```python
y2=3*x
```

```python
#Line plot
```

```python
plt.plot(x,y1)
plt.show()
```

```python
fig=plt.figure(figsize=(4,4))
```

```python
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)
plt.plot(x,y2,color='r',linestyle='-.',linewidth=3)
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

```python
#bar-plot
```

```python
student = {"Bob":87,"Matt":56,"Sam":27}
```

```python
names = list(student.keys())
values = list(student.values())
```

```python
plt.bar(names,values)
plt.title("Bar Plot")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
plt.barh(names,values,color='g')
plt.title("Bar Plot")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
#scatter-plot
```

```python
x=np.arange(1,11)
y1=2*x
y2=3*x
y=y1+y2
plt.scatter(x,y)
plt.title("Scatter Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

```python
x=[8,1,6,3,8,1,5]
y1=[6,2,7,9,1,4,5]
y2=[6,2,7,9,1,4,5]
```

```python
x=np.arange(1,11)
y1=2*x
y2=3*x

plt.subplot(1,2,1)
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)

plt.subplot(1,2,2)
plt.plot(x,y2,color='r',linestyle=':',linewidth=2)

plt.show()
```

```python
x=[10,20,30,40,50,60,70,80,90]
a=[8,1,7,2,0,3,7,3,2]

plt.scatter(x,a,c="red")
plt.show()
```

```python
x=[10,20,30,40,50,60,70,80,90]
a=[8,1,7,2,0,3,7,3,2]

plt.scatter(x,a,c="g",marker="*",s=100)
plt.show()
```

```python
x=[10,20,30,40,50,60,70,80,90]
a=[8,1,7,2,0,3,7,3,2]
b=[7,2,5,6,9,1,4,5,3]

plt.scatter(x,a,marker="*",c="g",s=100)
plt.scatter(x,b,marker=".",c="r",s=200)
plt.show()
```

```python
x=[10,20,30,40,50,60,70,80,90]
a=[8,1,7,2,0,3,7,3,2]
b=[7,2,5,6,9,1,4,5,3]

plt.subplot(1,2,1)
plt.scatter(x,a,marker="*",c="g",s=100)

plt.subplot(1,2,2)
plt.scatter(x,b,marker=".",c="r",s=200)

plt.show()
```

```python
#pie
```

```python
fruit = ['Apple','Orange','Mango','Guava']
quantity = [67,34,100,29]
```

```python
plt.pie(quantity,labels=fruit)
plt.show()
```

```python
plt.pie(quantity,labels=fruit,autopct='%0.1f%%',colors=['violet','indigo','blue','green'])
plt.show()
```

```python
plt.pie(quantity,labels=fruit,radius=1)
plt.pie([1],colors=['w'],radius=0.5)
plt.show()
```

```python
plt.pie(quantity,labels=fruit,autopct='%0.1f%%',radius=1)
plt.pie([1],colors=['w'],radius=0.5)
plt.show()
```
